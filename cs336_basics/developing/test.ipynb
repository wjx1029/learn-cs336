```python
import regex as re
```

```python
with open('test.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip('\n')
        if not line:
            continue

        part = line.split(',')
        for p in part:
            print(p)
        
```

```python
text = "<|begin|>u don't have to be.\nscared of the<|begin|> loud dog, I'll protect you.\n<|endoftext|>"
special_tokens = ["<|endoftext|>", "<|begin|>"]
# 自动转义并拼接特殊标记
special_pattern = '|'.join(re.escape(token) for token in special_tokens)
pattern = f"{special_pattern}|" + r"'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+"
parts = re.findall(pattern, text)
for part in parts:
    print([b  for b in part.encode()])
```
